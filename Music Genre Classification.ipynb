# MODEL TRAINING AND BUILDING

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
df.columns
```

```python
df2=df.drop(['Artist Name','Track Name','mode','time_signature','Class','Collabration'],axis=1)
df2
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
df2=scaler.fit_transform(df2)
```

```python
df2=pd.DataFrame(df2,columns=[ 'Popularity', 'danceability', 'energy',
       'key', 'loudness',  'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'])
df2
```

```python
df=df.drop([ 'Popularity', 'danceability', 'energy',
       'key', 'loudness',  'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_in min/ms'],axis=1)
```

```python
df
```

```python
df2
```

```python
df=pd.concat([df,df2],axis=1)
```

```python
df
```

```python
list(df['Artist Name'].unique())
```

```python
len(df['Artist Name'].unique())
```

```python
df=df.drop(['Artist Name','Track Name'],axis=1)
```

```python
df
```

```python
Y=df['Class']
Y
```

```python
X=df.drop(['Class'],axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=77)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```


# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='multinomial')
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

#  SVM - STANDARD VECTOR CLASSIFIER



```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,Y_train)
```

```python
svc.score(X_test,Y_test)
```

# RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)
```

# K NEIGHBORS CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

```python
knn.predict(X_test)
```
